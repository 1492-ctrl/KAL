```python
import numpy as np


# ==========================
# Gram-Schmidt QR Decomposition
# ==========================
def qr_decomposition(A):
    n = len(A)

    Q = np.zeros((n, n), dtype=float)
    R = np.zeros((n, n), dtype=float)

    for j in range(n):

        v = A[:, j].copy()

        for i in range(j):
            R[i][j] = np.dot(Q[:, i], A[:, j])
            v = v - (R[i][j] * Q[:, i])

        R[j][j] = np.linalg.norm(v)
        Q[:, j] = v / R[j][j]

    return Q, R


# ==========================
# QR Iteration (10x)
# ==========================
def qr_iteration(A, max_iter=10):

    Ak = A.astype(float)

    print("A0 =")
    print(Ak)
    print("=" * 40)

    for k in range(max_iter):

        # QR decomposition
        Q, R = qr_decomposition(Ak)

        # Ak+1 = RQ
        Ak = np.dot(R, Q)

        print("Iterasi ke-", k + 1)

        print("Q =")
        print(Q)

        print("R =")
        print(R)

        print("A" + str(k + 1) + " =")
        print(Ak)

        print("=" * 40)

    return Ak


def main():

    A = np.array([
        [2, 1],
        [1, 2]
    ], dtype=float)

    hasil = qr_iteration(A, 10)

    print("\nMatriks akhir:")
    print(hasil)

    print("\nEigenvalue:")
    print(np.diag(hasil))


if __name__ == "__main__":
    main()
```
