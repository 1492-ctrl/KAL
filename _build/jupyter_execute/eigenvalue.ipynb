# Tugas EigenValue

## menggunakan desimal

```python
import numpy as np

def qr_decomposition(A):
    n = len(A)

    Q = np.zeros((n, n))
    R = np.zeros((n, n))

    for j in range(n):

        v = A[:, j].copy()

        for i in range(j):
            R[i, j] = np.dot(Q[:, i], A[:, j])
            v = v - R[i, j] * Q[:, i]

        R[j, j] = np.linalg.norm(v)
        Q[:, j] = v / R[j, j]

    return Q, R

def qr_iteration(A, jumlah_iterasi=10):

    Ak = A.astype(float)

    print("A0 =")
    print(Ak)
    print()

    for k in range(jumlah_iterasi):

        print("=" * 40)
        print("ITERASI", k + 1)

        # QR decomposition
        Q, R = qr_decomposition(Ak)

        print("\nQ =")
        print(Q)

        print("\nR =")
        print(R)

        # Ak+1 = RQ
        Ak = R @ Q

        print("\nA{} =".format(k + 1))
        print(Ak)
        print()

    return Ak
```

```python
A = np.array([
    [2, 1],
    [1, 2]
], dtype=float)

hasil = qr_iteration(A, 10)

print("=" * 40)
print("MATRIKS AKHIR")
print(hasil)

print("\nEIGENVALUE:")
print(np.diag(hasil))
```

## menggunakan pecahan

```python
import sympy as sp

def qr_decomposition(A):

    n = A.rows

    Q = sp.zeros(n, n)
    R = sp.zeros(n, n)

    for j in range(n):

        v = A[:, j]

        for i in range(j):

            R[i, j] = (Q[:, i].T * A[:, j])[0]

            v = v - R[i, j] * Q[:, i]

        R[j, j] = sp.sqrt((v.T * v)[0])

        Q[:, j] = v / R[j, j]

    return Q, R

def qr_iteration(A, jumlah_iterasi=10):

    Ak = A

    print("A0 =")
    sp.pprint(Ak)
    print()

    for k in range(jumlah_iterasi):

        print("=" * 40)
        print("ITERASI", k + 1)

        Q, R = qr_decomposition(Ak)

        print("\nQ =")
        sp.pprint(Q)

        print("\nR =")
        sp.pprint(R)

        # Ak+1 = RQ
        Ak = sp.simplify(R * Q)

        print("\nA{} =".format(k + 1))
        sp.pprint(Ak)
        print()

    return Ak

```

```python
A = sp.Matrix([
    [2, 1],
    [1, 2]
])

hasil = qr_iteration(A, 10)

print("=" * 40)
print("MATRIKS AKHIR")
sp.pprint(hasil)

print("\nEIGENVALUE:")
sp.pprint(hasil.diagonal())
```
