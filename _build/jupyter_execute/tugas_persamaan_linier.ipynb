# Tugas Persamaan Linier

1. Apa itu persamaan linier?
Persamaan linier adalah persamaan matematika yang variabelnya berpangkat satu, sehingga jika digambarkan pada grafik koordinar kartesius akan membentuk garis lurus. Persamaan ini tidak memiliki variabel dengan pangkat lebih dari satu atau perkalian antarvariabel. dengan bentuk umum 

2. Buat dua persamaan linier dengan dua variabel

Persamaan pertama: 2x + y = 4

Persamaan kedua: x - y = 1
